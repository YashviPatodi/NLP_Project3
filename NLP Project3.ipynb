```python
import pandas as pd
titanic_df= pd.read_csv('titanic.csv')
titanic_df.info()
print(titanic_df.head())
```

```python
#Checking for discrepancies in the data types
print(titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId']].dtypes)

titanic_df= titanic_df.dropna()
#Modify data types if necessary- For Example as follows:
titanic_df['Fare']= titanic_df['Fare'].astype(int)
print(titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId']].dtypes)

titanic_df['Parch']= titanic_df['Parch'].astype(float)
print(titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId']].dtypes)
```

```python
stats= titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId', 'Age', 'SibSp']].agg(['mean', 'median', 'std', 'min', 'max', 'quantile'])
#2nd_quartile and median are the same
stats.loc['1st_quartile']= titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId', 'Age', 'SibSp']].quantile(0.25)
stats.loc['2nd_quartile']= titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId', 'Age', 'SibSp']].quantile(0.5)
stats.loc['3rd_quartile']= titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId', 'Age', 'SibSp']].quantile(0.75)
stats.loc['4th_quartile']= titanic_df[['Fare', 'Parch', 'Survived', 'PassengerId', 'Age', 'SibSp']].quantile(1)
print(stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

variables=['Age', 'Fare', 'Parch', 'SibSp', 'Survived']

plt.figure(figsize=(15,10))
for i, var in enumerate(variables, 1):
  plt.subplot(3,2,i)
  sns.histplot(titanic_df[var].dropna(), kde=True)
  plt.title(f'Distribution of {var}')
  plt.xlabel(var)
  plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Binomial Distribution

import numpy as np
import matplotlib.pyplot as plt

n,p= 10,0.5
data= np.random.binomial(n,p, 1000)

plt.hist(data, bins=10, density= True, alpha=0.6, color='g')
#alpha is transperancy of the graph
#value of alpha lies between 0 and 1
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

```python
# Poisson Distribution

import numpy as np
import matplotlib.pyplot as plt

lam=3
data= np.random.poisson(lam, 1000)

plt.hist(data, bins=15, density= True, alpha=0.6, color='b')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

```python
# Hypergeometric Distribution

import numpy as np
import matplotlib.pyplot as plt

#Parameters: N= total population,
#K= total success states, n=number of draws
N,K,n= 20,7,12
data= np.random.hypergeometric(K, N-K, n, 1000)

plt.hist(data, bins=range(n+1), density= True, alpha=0.6, color='r')
plt.title('Hypergeometric Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

```python
# Geometric Distribution

import numpy as np
import matplotlib.pyplot as plt

#Parameters: p= probability of success
p= 0.2
data= np.random.geometric(p, 1000)

plt.hist(data, bins=30, density= True, alpha=0.6, color='m')
plt.title('Geometric Distribution')
plt.xlabel('Number of Trials until First Success')
plt.ylabel('Probability')
plt.show()
```

```python
import pandas as pd
titanic_df= pd.read_csv('titanic.csv')
titanic_df.info()
print(titanic_df.head())
```

```python
def detect_outliers(data, variable):
  Q1= data[variable].quantile(0.25)
  Q2= data[variable].quantile(0.75)
  IQR= Q2-Q1
  lower_bound= Q1-1.5*IQR
  upper_bound= Q2+1.5*IQR
  outliers= data[(data[variable]< lower_bound) | (data[variable]>upper_bound)]
  return outliers

outliers_age= detect_outliers(titanic_df, 'Age')
outliers_fare= detect_outliers(titanic_df, 'Fare')

print("Outliers in Age:")
print(outliers_age[['Age']])
print("\nOutliers in Fare:")
print(outliers_fare[['Fare']])

titanic_df_cleaned= titanic_df[~titanic_df.index.isin(outliers_age.index)]
titanic_df_cleaned= titanic_df_cleaned[~titanic_df_cleaned.index.isin(outliers_fare.index)]

print(titanic_df_cleaned)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
#Parameters
mu,sigma=0,1 #mean and standard deviation
#Generate Data
data= np.random.normal(mu, sigma, 1000)

plt.figure(figsize=(10,6))
sns.histplot(data, kde=True, stat="density", linewidth=0)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
#Uniform Distribution
from scipy. stats import uniform
#Parameters
loc, scale= 0,1 #lower bound and range

data= np.random.uniform(loc, scale, 1000)

plt.figure(figsize=(10,6))
sns.histplot(data, kde=False, stat="density", linewidth=0)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import lognorm
mean=0
sigma=1
data=np.random.lognormal(mean, sigma, 1000)
plt.figure(figsize=(10,6))
sns.histplot(data, kde=True, stat="density", linewidth=0)
plt.title('Lognormal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.xlim(0, max(data))
plt.show()
```

```python
#F Distribution
from scipy.stats import f
dfn, dfd= 5,2
data=np.random.f(dfn, dfd, 1000)
plt.figure(figsize=(10,6))
sns.histplot(data, kde=True, stat="density", linewidth=0)
plt.title('F Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
from scipy.stats import chi2
#Parameters
df=2 # degrees of freedom
data=np.random.chisquare(df,1000)
plt.figure(figsize=(10,6))
sns.histplot(data, kde=True, stat="density", linewidth=0)
plt.title('Chi-Square Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
#Exponential Distribution
from scipy.stats import expon
scale=1
data= np.random.exponential(scale, 1000)
plt.figure(figsize=(10,6))
sns.histplot(data, kde=True, stat="density", linewidth=0)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
